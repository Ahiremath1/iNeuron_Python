## 2:Numbers ranging from 2000 to 3200, divisible by 7 and not multiples of 5 

```python

results = list()
for numbers in list(range(2000, 3201)):
    if numbers%7==0 and numbers%5 !=0:
        results.append(numbers)
        
print(results)
```

   ## 4. Program to find volume of sphere given the diameter

```python
import math
diameter = input("Enter diameter of spehre in cm:")
radius = float(diameter)/2
Volume = (4/3)*math.pi*(radius)**3
print("Volume in cubic centimeters:",round(Volume,2))
```

## 2: Create a patter of using nested loop:

```python
lines = input("enter number of rows:")
line = int(lines)
for i in range(0,line):
    for ii in range(0,i+1):
        print('*', end='')
    print("\r")
for i in range(line, 0,-1):
    for j in range(0,i-1):
        print('*', end='')
    print("\r")
```
